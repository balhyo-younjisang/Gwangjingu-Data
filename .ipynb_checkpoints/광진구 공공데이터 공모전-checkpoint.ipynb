# 광진구 내 어린이공원 공백 지역 식별 및 분석

```python
import geopandas as gpd
from shapely.geometry import Point
import csv
from sklearn.cluster import KMeans
import numpy as np
import folium

# GeoDataFrame으로 변환
park_gdf = gpd.GeoDataFrame(
    child_park_df,
    geometry=gpd.points_from_xy(child_park_df['경도'], child_park_df['위도']),
    crs="EPSG:4326"
)

# 좌표계를 거리 단위로 변환
park_gdf = park_gdf.to_crs(epsg=5181)

# 200m 버퍼 생성
park_gdf['buffer'] = park_gdf.geometry.buffer(200)

jumin_df = pd.read_csv("./data/광진구_연령별_등록인구.csv", quoting=csv.QUOTE_NONE)
# 컬럼 이름 정리 (큰따옴표 제거)
jumin_df.columns = jumin_df.columns.str.replace('"', '').str.strip()

print(jumin_df[['동별', '연령별', '항목', '2024. 4/4']].head(20))
# 필요한 컬럼만 선택
jumin_df = jumin_df[['동별', '연령별', '항목', '2024. 4/4']]

# 따옴표 제거
jumin_df['동별'] = jumin_df['동별'].str.replace('"', '').str.strip()
jumin_df['연령별'] = jumin_df['연령별'].str.replace('"', '').str.strip()
jumin_df['항목'] = jumin_df['항목'].str.replace('"', '').str.strip()

jumin_df['2024. 4/4'] = jumin_df['2024. 4/4'].replace("-", None)
jumin_df['2024. 4/4'] = jumin_df['2024. 4/4'].str.replace('"', '').str.strip()
jumin_df['2024. 4/4'] = pd.to_numeric(jumin_df['2024. 4/4'], errors='coerce')

# (중요) 조건 확인
df_filtered = jumin_df[
    ((jumin_df['연령별'] == '0~4세') | (jumin_df['연령별'] == '5~9세')) &
    (jumin_df['항목'] == '한국인')
]

# 필터링 결과 확인
print(df_filtered.head())

# 동별로 합산
dong_child_population = df_filtered.groupby('동별')['2024. 4/4'].sum().reset_index()

# 컬럼 이름 변경
dong_child_population.columns = ['동', '유아_인구']
dong_child_population = dong_child_population[dong_child_population['동'] != '광진구']
dong_child_population['동'] = dong_child_population['동'].str.replace("제\d동", "동", regex=True)

print(dong_child_population)

gwangjingu_dong_coords_df = pd.read_csv("./data/행정_법정동 중심좌표.CSV", encoding="cp949")
# gwangjingu_dong_coords_df = gwangjingu_dong_coords_df.dropna(axis=1)

gwangjingu_dong_coords_df = gwangjingu_dong_coords_df[['시군구', '읍면동', '위도', '경도']].dropna()
gwangjingu_dong_coords_df = gwangjingu_dong_coords_df[(gwangjingu_dong_coords_df['시군구'] == '광진구')]
gwangjingu_dong_coords_df

# 행정동 기준 광진구 내 동별 중심 좌표
dong_coords_df = {
    '화양동': [37.543250, 127.073393],
    '군자동': [37.553038, 127.073657],
    '중곡제1동': [37.538210, 127.082191],
    '중곡제2동': [37.538210, 127.082191],
    '중곡제3동': [37.538210, 127.082191],
    '중곡제4동': [37.538210, 127.082191],
    '능동': [37.550578, 127.081722],
    '광장동': [37.547835, 127.104695],
    '자양제1동': [37.538210, 127.082191],
    '자양제2동': [37.538210, 127.082191],
    '자양제3동': [37.538210, 127.082191],
    '자양제4동': [37.538210, 127.082191],
    '구의제1동': [37.538210, 127.082191],
    '구의제2동': [37.538210, 127.082191],
    '구의제3동': [37.538210, 127.082191],
}

dong_coords_df = pd.DataFrame([
    {"동": k.replace("제", ""), "위도": v[0], "경도": v[1]} for k, v in dong_coords_df.items()
])


# 병합
merged_df = dong_coords_df.merge(dong_child_population, on='동', how='left')
merged_df

# 인구 지점 생성
gwangjin_geo = gpd.GeoDataFrame(
    merged_df,
    geometry=gpd.points_from_xy(merged_df['경도'], merged_df['위도']),
    crs="EPSG:4326"
).to_crs(epsg=5181)

# 공원 버퍼 통합 (멀티폴리곤)
all_buffers = park_gdf['buffer'].union_all()

# 공원 커버 범위에 포함되지 않는 인구지점 필터링
gwangjin_uncovered = gwangjin_geo[~gwangjin_geo.geometry.within(all_buffers)]

# 좌표만 추출
coords = np.array(list(zip(gwangjin_uncovered.geometry.x, gwangjin_uncovered.geometry.y)))
coords = np.unique(coords, axis=0) 

# 클러스터링 (예: 5개 신규 공원 입지)
kmeans = KMeans(n_clusters=2, random_state=0).fit(coords)

# 중심 좌표 추출
centroids = kmeans.cluster_centers_
centers_df = pd.DataFrame(centroids, columns=['x', 'y'])

# GeoDataFrame으로 변환
centers_gdf = gpd.GeoDataFrame(
    centers_df,
    geometry=gpd.points_from_xy(centers_df['x'], centers_df['y']),
    crs="EPSG:5181"
)

# 위경도 좌표계로 변환
centers_gdf = centers_gdf.to_crs(epsg=4326)

# 중심 좌표 기반 지도
m = folium.Map(location=[lat, lon], zoom_start=13)

# 기존 공원 표시
for _, row in park_df.iterrows():
    folium.Circle(
        location=[row['위도'], row['경도']],
        radius=200,
        color="blue",
        fill=True,
        fill_opacity=0.3,
    ).add_to(m)
    
for idx, row in centers_gdf.iterrows():
    folium.Marker(
        location=[row.geometry.y, row.geometry.x],
        popup=f"신규 공원 후보지 {idx+1} [{row.geometry.y}, {row.geometry.x}]",
        icon=folium.Icon(color="green", icon="tree")
    ).add_to(m)

m
```

```python
import pandas as pd
from scipy.spatial import cKDTree
from sklearn.neighbors import BallTree

# 광진구 범위 내 격자 생성
lats = np.arange(37.515, 37.6, 0.001)  # 약 100m
lons = np.arange(123.7, 127.12, 0.001)

grid_points = [(lat, lon) for lat in lats for lon in lons]
candidate_df = pd.DataFrame(grid_points, columns=['lat', 'lon'])

# 반경 내 어린이집 계산
carecenter_df = pd.read_csv("./data/서울시 광진구 어린이집 정보.csv" , encoding='cp949')
carecenter_df = carecenter_df[['어린이집명', '시설 위도(좌표값)', '시설 경도(좌표값)', '아동수-총계']].dropna()
print(carecenter_df.head())
carecenter_df_radians = np.radians(carecenter_df[['시설 위도(좌표값)', '시설 경도(좌표값)']])
tree = BallTree(carecenter_df_radians, metric='haversine')
radius_km = 0.2 / 6371  # 200m

candidate_df['nearby_childcare'] = tree.query_radius(np.radians(candidate_df[['lat', 'lon']]), r=radius_km, count_only=True)
candidate_df
```
